```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as pt
from scipy import ndimage

a=Image.open("images.jpg")
b=np.array(a)
c=ndimage.rotate(b,60,cval=10)
x=Image.fromarray(c)
pt.imshow(x)
pt.show()
```

```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as pt
from scipy import ndimage

a=Image.open("sample.jpg")
b=np.array(a)
c=ndimage.zoom(b,(1,1,1))
x=Image.fromarray(c)
pt.imshow(x)
pt.show()
```

```python
#shearing x on y
import numpy as np
from PIL import Image
from matplotlib import pyplot as pt
from scipy import ndimage
a=Image.open("images.jpg")
b=np.array(a)
f=[[1,0.5,0],[0,1,0],[0,0,1]]
c=ndimage.affine_transform(b,f)
x=Image.fromarray(c)
pt.imshow(x)
pt.show()
```

```python
#shearing y on x
import numpy as np
from PIL import Image
from matplotlib import pyplot as pt
from scipy import ndimage
a=Image.open("images.jpg")
b=np.array(a)
f=[[1,0,0],[0.5,1,0],[0,0,1]]
c=ndimage.affine_transform(b,f)
x=Image.fromarray(c)
pt.imshow(x)
pt.show()
```
