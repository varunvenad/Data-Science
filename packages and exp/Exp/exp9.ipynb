```python
#12. Programs on feedforward network to classify any standard dataset available in the public  domain. 
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
irisData=load_iris ()
x=irisData.data
y=irisData.target
print("\n------------------------\ndata and target\n------------------------\n")
print(x)
print("\n")
print(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
clt=MLPClassifier(random_state=1,activation='logistic',max_iter=300).fit(x_train,y_train)
print("\n--------------------------------------\npredict probability of test data targets \n--------------------------------------\n")
e=clt.predict_proba(x_test)
print(e)
print("\n------------------------\npredict test data targets\n------------------------\n")
l=clt.predict(x_test)
k=clt.score(x_test,y_test)
print(k)
```
